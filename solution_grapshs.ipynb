## Load the Data

```python
import pandas as pd
data = pd.read_csv('World-Stock-Prices-Dataset.csv')
print(data.head())
```

### 1. Line Graph: Average Closing Prices for the Top 5 Stocks Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Date': pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02']),
    'Brand_Name': ['Stock A', 'Stock A', 'Stock B', 'Stock B'],
    'Close': [100, 105, 200, 195]
})

highest_avg_closing_price = data.groupby('Brand_Name')['Close'].mean().sort_values(ascending=False)

top_5_stocks = highest_avg_closing_price.head(5).index

top_5_data = data[data['Brand_Name'].isin(top_5_stocks)]

avg_closing_prices = top_5_data.groupby(['Date', 'Brand_Name'])['Close'].mean().unstack()

avg_closing_prices.plot(figsize=(10, 6))
plt.title('Average Closing Prices for Top 5 Stocks Over Time')
plt.xlabel('Date')
plt.ylabel('Average Closing Price')
plt.legend(title='Stock')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Ensure labels fit into the figure
plt.show()
```

## 2. Bar Graph: Trading Volume by Industry

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Industry_Tag': ['Tech', 'Tech', 'Finance', 'Finance', 'Health', 'Health'],
    'Volume': [1000, 1500, 2000, 2500, 3000, 3500]
})

industry_volume = data.groupby('Industry_Tag')['Volume'].sum()

industry_volume.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Trading Volume by Industry')
plt.xlabel('Industry')
plt.ylabel('Total Volume')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## 3. Scatter Plot: Link Between Volume and Price Changes

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Volume': [1000, 1500, 2000, 2500, 3000, 3500],
    'Price_Change_Percent': [5.2, 3.4, 4.5, 6.1, 2.8, 3.0]
})

plt.figure(figsize=(10, 6))
plt.scatter(data['Volume'], data['Price_Change_Percent'], alpha=0.5, color='purple')

plt.title('Link Between Volume and Price Changes')
plt.xlabel('Volume')
plt.ylabel('Price Change (%)')

plt.grid()

plt.show()
```

## 4. Box Plot: Comparing Price Changes Across Industries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.DataFrame({
    'Industry_Tag': ['Tech', 'Tech', 'Finance', 'Finance', 'Health', 'Health', 'Tech', 'Finance', 'Health'],
    'Price_Change_Percent': [5.2, 3.4, 4.5, 6.1, 2.8, 3.0, 7.1, 8.0, 4.2]
})

plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Industry_Tag', y='Price_Change_Percent')

plt.title('Price Changes Across Industries')
plt.xlabel('Industry')
plt.ylabel('Price Change (%)')

plt.xticks(rotation=45)

plt.show()
```

## 5. Line Graph: Stock Prices for One Industry Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Date': pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02']),
    'Industry_Tag': ['technology', 'technology', 'finance', 'finance', 'technology', 'technology'],
    'Close': [150, 155, 200, 195, 160, 165]
})

tech_stocks = data[data['Industry_Tag'] == 'technology']

tech_trend = tech_stocks.groupby('Date')['Close'].mean()

tech_trend.plot(figsize=(10, 6), color='green')
plt.title('Technology Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Average Closing Price')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to ensure everything fits
plt.show()
```

### 6. Heatmap: Relationships Between Stock Prices by Industry or Country

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.DataFrame({
    'Industry_Tag': ['Technology', 'Technology', 'Finance', 'Finance', 'Technology', 'Healthcare'],
    'Country': ['USA', 'Canada', 'USA', 'Canada', 'USA', 'Canada'],
    'Close': [150, 155, 200, 195, 160, 165]
})

heatmap_data = data.pivot_table(values='Close', index='Industry_Tag', columns='Country', aggfunc='mean')

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='.1f')

plt.title('Stock Prices by Industry and Country')
plt.xlabel('Country')
plt.ylabel('Industry')

plt.show()
```

## 7. Histogram: Dividend Payouts for All Stocks

## 8. Pie Chart: Stocks Grouped by Country

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Country': ['USA', 'Canada', 'USA', 'Canada', 'USA', 'Germany', 'USA', 'Germany', 'Canada', 'Germany']
})

country_counts = data['Country'].value_counts()

country_counts.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', startangle=140, cmap='Set3')

plt.title('Stocks Grouped by Country')

plt.ylabel('')
plt.show()
```

## 9. Line Graph: Price Trends After Stock Splits

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Date': pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02']),
    'Stock Splits': [0, 1, 0, 0, 1, 0],
    'Close': [150, 155, 160, 158, 162, 165]
})

stock_splits = data[data['Stock Splits'] > 0]

price_trend = stock_splits.groupby('Date')['Close'].mean()

price_trend.plot(figsize=(10, 6), color='blue')
plt.title('Price Trends After Stock Splits')
plt.xlabel('Date')
plt.ylabel('Average Closing Price')
plt.grid(True)
plt.show()
```

### 10. Stacked Bar Graph: Closing Prices Grouped by Industry and Country

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Industry_Tag': ['Technology', 'Technology', 'Finance', 'Finance', 'Healthcare', 'Healthcare'],
    'Country': ['USA', 'Canada', 'USA', 'Canada', 'USA', 'Canada'],
    'Close': [150, 155, 200, 195, 160, 165]
})

stacked_data = data.pivot_table(values='Close', index='Industry_Tag', columns='Country', aggfunc='mean')

stacked_data.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')

plt.title('Closing Prices Grouped by Industry and Country')
plt.xlabel('Industry')
plt.ylabel('Average Closing Price')

plt.xticks(rotation=45)
plt.legend(title='Country')
plt.show()
```
